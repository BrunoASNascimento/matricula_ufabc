```
df_personal = pd.DataFrame(data)
df_personal = df_personal.loc[[('Repr' not in x) for x in df_personal['situacao']]]
df_personal.sort_values(['ano', 'disciplina'], ascending=[1, 0])
```

```
subjects = pd.read_excel('/content/drive/My Drive/ML/ufabc/data/categorias_catalogo_disciplinas_graduacao_2018_2019_V2.xlsx')
subjects.head()
```

```
cursos = subjects['Cursos/Categoria'].drop_duplicates().str.split(';', expand=True)
df_cursos = pd.Series(name='cursos')
for column in cursos:
    df_edit=cursos[column].dropna().rename('cursos')    
    df_cursos= pd.concat([df_cursos, df_edit], ignore_index=True).dropna()
df_cursos = df_cursos.drop_duplicates().sort_values()
display(df_cursos)
```

```
filter_str= 'BC&T - Bacharelado em Ciência e Tecnologia (OBR)'
# filter_str= 'BBT - Bacharelado em Biotecnologia (OBR)'
# filter_str = 'BN - Bacharelado em Neurociência (OBR)'
# filter_str='EENE - Engenharia de Energia (OBR)'
# filter_str='EAER Engenharia Aeroespacial (OBR)'
todo = subjects.loc[[(filter_str in x) for x in subjects['Cursos/Categoria']]]
todo = todo[~todo['Sigla'].isin(df_personal['codigo'])]
todo
```

```
disciplinas = pd.read_excel('/content/drive/My Drive/ML/ufabc/data/catalogo_disciplinas_graduacao_2018_2019.xlsx')
disciplinas_todo = disciplinas[disciplinas['SIGLA'].isin(todo['Sigla'])]
disciplinas_todo
```

```
# df_personal
df_finally = pd.DataFrame()
for disciplina in df_personal['disciplina'].values:   
    # df_finally = df_finally.append(disciplinas_todo.loc[[((disciplina.lower()).replace(' ','') in (x.lower()).replace(' ','')) for x in disciplinas_todo['RECOMENDAÇÃO']]])
    df_edit =disciplinas_todo.loc[[((disciplina.lower()).replace(' ','') in (x.lower()).replace(' ','')) for x in disciplinas_todo['RECOMENDAÇÃO']]]
    if df_edit.shape[0]==0:
        df_edit=disciplinas_todo
        df_edit['afinidade']=0
    else:
        df_edit['afinidade']=1
    
    df_finally = df_finally.append(df_edit)



df_finally =df_finally.groupby(['SIGLA','DISCIPLINA','TPI','RECOMENDAÇÃO'],as_index=False)['afinidade'].sum()
df_finally['afinidade'] = (df_finally['afinidade']/df_finally['RECOMENDAÇÃO'].str.split(';').str.len()).round(2)
df_finally[['T','P','I']] = df_finally['TPI'].str.split('-', expand=True)
df_finally['creditos'] = df_finally['T'].astype('float')+df_finally['P'].astype('float')
df_finally.loc[df_finally['RECOMENDAÇÃO'].str.match('^Não.*')== True,'afinidade']=1
df_finally=df_finally.sort_values(by=['afinidade','creditos'],ascending=[False,True] )
display(df_finally)
```
